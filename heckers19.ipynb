```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string

import warnings
warnings.filterwarnings("ignore")

```

```python
df = pd.read_csv('/kaggle/input/truth-seekers/train.csv')
df.head(10)
```

```python
df['label'].value_counts()
```

```python
df=df.sample(frac=1)
```

```python
from tqdm import tqdm
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess_text(text_data):
    preprocessed_text = []
      
    for sentence in tqdm(text_data):
        sentence = re.sub(r'[^\w\s]', '', sentence)
        preprocessed_text.append(' '.join(token.lower()
                                  for token in str(sentence).split()
                                  if token not in stopwords.words('english')))
  
    return preprocessed_text

df['text'] = preprocess_text(df['text'].values)
```

```python
x=df['text']
y=df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.linear_model import LogisticRegression, PassiveAggressiveClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier

from imblearn.pipeline import make_pipeline as make_pipeline_imb
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling  import RandomOverSampler

model = make_pipeline_imb(TfidfVectorizer(), RandomUnderSampler(),
                          PassiveAggressiveClassifier(max_iter = 5000,validation_fraction=0.2,verbose =1,n_iter_no_change=100))

model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
test=pd.read_csv('/kaggle/input/truth-seekers/test.csv')
test.head(10)
```

```python
test['text'] = preprocess_text(test['text'].values)
```

```python
test['label']=model.predict(test['text'])
```
